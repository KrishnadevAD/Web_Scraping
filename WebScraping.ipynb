**WEB SCRAPPING**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.worldometers.info/coronavirus/"
```

```python
response=requests.get(url).content.decode()
```

```python
soup=BeautifulSoup(response,"html.parser")
```

```python
world_data=soup.find("tbody").find_all("tr")
```

```python
complete_data=list()
for i in range(8,len(world_data)):
  data=[]
  list_data=world_data[i].find_all("td")
  for i in list_data:
    data.append(i.text)
    complete_data.append(data)
```

```python
complete_data
```

```python
mapped_data=list(map(lambda x:x[1:10] + [x[12]] + [x[14]],complete_data))
```

```python
column_names=[" Country Name",
              "Total Cases",
              "New Cases",
              "Total Deaths",
              "New Deaths ",
              "Total Recovered",
              "New Recovered",
              "Active Cases",
              "Serious Cases",
              "Total Tests",
              "Population"]
```

```python
mapped_data
```

```python

df=pd.DataFrame(mapped_data,columns=column_names)
df.to_csv("covid_data aug 27.csv",index=False)
```
